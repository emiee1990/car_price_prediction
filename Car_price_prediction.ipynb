# Handling Missing Data

```python
from sklearn.impute import SimpleImputer
```

```python
num_imputer = SimpleImputer(strategy='median')
cat_imputer = SimpleImputer(strategy='most_frequent')

num_imputer.fit(df[num_col])
cat_imputer.fit(df[cat_col])
```

```python
df[num_col] = num_imputer.transform(df[num_col])
df[cat_col] = cat_imputer.transform(df[cat_col])
```

```python
df.isna().sum()
```

# Feature Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit(df[cat_col])
```

```python
enc_col = ohe.get_feature_names()
```

```python
enc_col
```

```python
temp_train = pd.DataFrame(data = ohe.transform(df[cat_col]).toarray(),columns = enc_col)
```

```python
temp_train
```

```python
temp_train.columns
```

```python
# final_cols = num_col + list(ohe.get_feature_names())
```

```python
df_final = pd.concat([df[num_col],temp_train,selling_price], axis=1)
```

```python
df_final.columns
```

```python
X = df_final.drop('Selling_Price',axis = 1)
y = df_final['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 32)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()
```

```python
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```
